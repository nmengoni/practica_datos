Lectura de informe de datos

```python
import sqlite3
import pandas as pd
```

```python
#Conexión a la base de datos
conn=sqlite3.connect("northwind.db")
```

```python
# Para ver las tablas vamos a usar una función de pandas que se llama read_sql 
tablas=pd.read_sql("""SELECT name, type FROM sqlite_master WHERE type IN ('table','view') ORDER BY name""", conn)
print(tablas)
```

Table vs View

Table: son ls datos de las  tablas originales

View: es cuando se guarda una quary especifica, tipo favorito porque lo consultas siempre

```python
# Guardar una lista con los nombres de las tablas unicamente
lista_tablas=tablas[tablas["type"]=="table"]["name"].tolist()
lista_tablas

#Ordenar alfabéticamente las tablas: por defecto ya están ordenadas alfabéticamente.
lista_tablas.sort(reverse=False)
lista_tablas
```

Para las columnas de cada, los tipos y las primary key

PRAGMA table_info --> Cuando lo ejecutivo devuelve una tabla con columnas que tiene nombres técnicos. Por defecto me devuelve la tabla con las columas: cid, names, type, notnull, dflt_value_pk

```python
def ver_columnas(tablas):
    df=pd.read_sql(f'PRAGMA table_info("{tablas}")', conn)
    df.columns=['cid','nombre','tipo','not_null','defalult','es_pk'] #renombrando las columnas
    df['es_pk']=df['es_pk'].apply(lambda x: 'Sí' if x==1 else 'No' ) #cambiando el valor de la columna es_pk a Sí o No
    df['not_null']=df['not_null'].apply(lambda x: 'No null' if x==1 else 'Null' ) 
    return df [['nombre','tipo','not_null','es_pk']] #seleccionando solo las columnas que me int
for tabla in lista_tablas:
    print(f'\n[tabla.upper()]')
    print(ver_columnas(tabla).to_string(index=False))
```
